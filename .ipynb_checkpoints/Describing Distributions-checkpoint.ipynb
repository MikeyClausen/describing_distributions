# Imports

```python
#Imports
import os
os.environ["OMP_NUM_THREADS"] = '1'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
url='https://docs.google.com/spreadsheets/d/1APV3pXiAszS_0mSgkiEt9IUNH-QmyX7KwxSAwuADl6Y/gviz/tq?tqx=out:csv&sheet=medical_data'
df=pd.read_csv(url)
df.head()
```

```python
df.columns
```

# VitD_levels

```python
df['VitD_levels'].value_counts()
```

```python
col='VitD_levels'
feature=df[col]
mean=feature.mean()
median=feature.median()
std=feature.std()
plus1std=mean+std
minus1std=mean-std
```

```python
sns.scatterplot(feature)
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.histplot(feature,ax=ax,stat='density',kde=True)
ax.axvline(mean,color='red',label=f'Mean={mean:,.2f}')
ax.axvline(median,color='green',ls='--',label=f'Median={median:,.2f}')
ax.axvline(plus1std,color='black',label=f'+1 std={plus1std:,.2f}')
ax.axvline(minus1std,color='black',label= f'-1 std={minus1std:,.2f}')
ax.axvspan(plus1std,minus1std,color='yellow',zorder=0)
ax.set_title(col)
ax.legend();
```

```python
kurt=stats.kurtosis(feature,fisher=False)
kurt
```
